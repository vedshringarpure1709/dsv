```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Iris.csv")
data.head(5)
```

```python
data['variety'].value_counts()
```

```python
data['sepal.length'].plot(kind = 'hist')
plt.show()
```

```python
data.plot(kind = "scatter", x = "sepal.length", y = "sepal.width")
```

```python
sns.jointplot(x = 'sepal.length', y = 'sepal.width', data = data, size = 5)
```

```python
sns.FacetGrid(data, hue = 'variety', palette = 'husl', height = 5) \
.map(plt.scatter, 'sepal.length', 'sepal.width') \
.add_legend()
```

```python
sns.FacetGrid(data, hue = 'variety', palette = 'husl', height = 5) \
.map(plt.scatter, 'petal.length', 'petal.width') \
.add_legend()
```

```python
sns.boxplot(x = 'variety', y = 'petal.length', palette = 'husl', data = data)
```

```python
sns.FacetGrid(data, hue = 'variety', palette = 'husl', height = 5) \
.map(sns.kdeplot, 'petal.length') \
.add_legend()
```

```python
data.boxplot(by = 'variety', figsize = (12, 10))
```

```python
sns.pairplot(data, hue = 'variety', size = 3)
plt.show()
```

```python
from pandas.plotting import parallel_coordinates
parallel_coordinates(data, 'variety', colormap = 'rainbow')
plt.show()
```
