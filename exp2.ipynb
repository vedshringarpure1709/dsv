```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.random import randint
import warnings
warnings.filterwarnings('ignore')
```

```python
test = randint(30, 70, 1000)
test
```

```python
import statistics
```

```python
mean = statistics.mean(test)
print("mean:", mean)
```

```python
variance = statistics.variance(test)
print("variance:", variance)
```

```python
median = statistics.median(test) 
print("median:", median)
```

```python
mode = statistics.mode(test)
print("mode", mode)
```

```python
standard_Deviation = np.std(test)
print("Standard Deviation:", standard_Deviation)
```

Implementation of normal distribution

```python
size = 10000
values = np.random.normal(mean, standard_Deviation, size)
```

```python
plt.hist(values, 100)
plt.axvline(values.mean())
plt.show()
```
